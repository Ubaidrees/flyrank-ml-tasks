## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/Ubaidrees/flyrank-ml-tasks/main/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(url)

print("--- 1. UNIT OF ANALYSIS ---")
print("Unit of Analysis: One content item (page) per 90-day snapshot window.")
print("\nSample Dataframe Structure:")
print(df[['content_id', 'client_id', 'clicks_90d', 'clicks_last_30d', 'trend_direction']].head(4))

print("\n--- 2. TARGET COLUMN SKETCH (is_drop_imminent) ---")
df['is_drop_imminent'] = (df['trend_direction'] == 'down').astype(int)
print(df[['content_id', 'clicks_last_30d', 'is_drop_imminent']].head(4))
print(f"\nTarget Class Distribution:\n{df['is_drop_imminent'].value_counts(normalize=True)}")
```

## 5. Why ML beats a fixed rule here
Why ML Beats a Fixed Rule: > A static business rule like "Flag page if position falls below 10" is too brittle. A page might drop in position but still sustain conversions due to seasonal intent shift, or conversely, stay at position 2 but lose 40% clicks because a competitor changed their meta title or layout.
A fixed rule cannot capture these multi-dimensional, non-linear signals. Machine learning beats a rule-based system by tracking complex multi-feature interactions simultaneously across thousands of URLs dynamically.
